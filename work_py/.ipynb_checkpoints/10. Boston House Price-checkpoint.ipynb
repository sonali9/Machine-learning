```python
from sklearn import model_selection
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(X,Y)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
test1 = np.loadtxt("gdtest1.csv",delimiter = ",")
test2 = np.loadtxt("prediction1.csv",delimiter = ",")
alg = LinearRegression()
alg.fit(x_train,y_train)
```

```python
y_pred = alg.predict(test1)
y_pred
```

```python

test3_score = alg.score(test1,y_pred)

print("Test score:", test3_score)
test1_score = alg.score(test1,test2)

print("Test score:", test1_score)
```
