```python
# customizing graphs
#plot function in x^3
import numpy as np
#x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
y=x**3
y2=x**2

#plt.plot(x,y,color="green",marker="p")
plt.plot(x,y,color="black",linewidth = 5,label="x^3")
plt.plot(x,y2,color="red",linewidth = 5,label="x^2")

plt.ylabel("y")
plt.xlabel("x")
plt.legend()
plt.title("Mathplotlib demo")

plt.show()
```

```python
import numpy as np
#color,marker,linewidth,labels for axis, datasets and the title,axis,grid,text
x=np.arange(0,5,0.1)
y=x**3
y2=x**2

plt.plot(x,y,color="black",linewidth = 5,label="x^3")
plt.plot(x,y2,color="red",linewidth = 5,label="x^2")

#plt.axis([0,10,0,200])

plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.text(2,80,"Text",fontsize=12)

plt.legend()
plt.title("Mathplotlib demo")

plt.show()
```

```python
#pie graphs
labels=["A","B","C","D"]
sizes = [3,4,6,2]
colors = ["blue","yellow","purple","pink"]
explode = [0.1,0,0,0]

plt.title("Split among classes")
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.2f%%",counterclock=False,startangle=90)

plt.axis("equal")
plt.show()
```
