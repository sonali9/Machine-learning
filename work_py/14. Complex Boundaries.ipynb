```python
from sklearn import datasets
import numpy as np
```

```python
boston = datasets.load_boston()
X = boston.data
Y = boston.target
print(X.shape)
print(Y.shape)
```

```python
import pandas as pd
df = pd.DataFrame(X,Y)
df.columns = boston.feature_names
df1 = pd.DataFrame(X,Y)
df1.columns = boston.feature_names

# df1.iloc[:,:] = df.iloc[:,:].copy()

for i in range(0,13):
    for j in range(i,13):
        x = df.columns[i]+"_"+df.columns[j]
        
        df1.loc[:,x]=df.loc[:,df.columns[i]]*df.loc[:,df.columns[j]]
        print(df1.loc[:,x],end = " ")

print()

df["AGE_AGE"]=df.AGE**2
print(df.describe())
df1.describe()
```

```python
X2 = df.values
print(X2.shape)
X3 = df1.values
print(X3.shape)
```

```python
from sklearn import model_selection
[xtr,xte,ytr,yte] = model_selection.train_test_split(X,Y, random_state=0)
[x2tr,x2te,y2tr,y2te] = model_selection.train_test_split(X2,Y,random_state=0)
[x3tr,x3te,y3tr,y3te] = model_selection.train_test_split(X3,Y,random_state=0)
print(xtr.shape)
print(xte.shape)
print(ytr.shape)
print(yte.shape)
print(x2tr.shape)
print(x2te.shape)
print(y2tr.shape)
print(y2te.shape)
print(x3tr.shape)
print(x3te.shape)
print(y3tr.shape)
print(y3te.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
alg1 = LinearRegression()
alg2 = LinearRegression()
alg3 = LinearRegression()
alg1.fit(xtr,ytr)
alg2.fit(x2tr,y2tr)
alg3.fit(x3tr,y3tr)
```

```python
ypred = alg1.predict(xte)
```

```python
from matplotlib import pyplot as plt
plt.scatter(ypred,yte)
plt.axis([0,40,0,40])
plt.show()
```

```python
train_score = alg1.score(xtr,ytr)
test_score = alg1.score(xte,yte)
print("Train score:", train_score)
print("Test score:", test_score)
```

```python
train2_score = alg2.score(x2tr,y2tr)
test2_score = alg2.score(x2te,y2te)
print("Train score:", train2_score)
print("Test score:", test2_score)
```

```python
train3_score = alg3.score(x3tr,y3tr)
test3_score = alg3.score(x3te,y3te)
print("Train score:", train3_score)
print("Test score:", test3_score)
```
