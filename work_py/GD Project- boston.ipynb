```python
import numpy as np
from sklearn import model_selection
```

```python
def step(points, learning_rate, m):
    n=len(points[0])-1
    m_slope = np.zeros(n+1)
    M = len(points)
    m[n]=1
    for i in range(M):
        x = points[i,0:n]
        y = points[i,n]
        for j in range(n):
            m_slope[j] += (-1/M) * ((y - (m[j] * x[j])) * x[j]) 
    new_m = m - learning_rate * m_slope
    return new_m
```

```python
def gd(points, learning_rate, no_iteration):
    m = np.zeros(len(points[0]))
    for i in range(no_iteration):
        m = step(points, learning_rate, m)
        print(i, "Cost : ",cost(points, m))
    return m
```

```python
def cost(points, m):
    total_cost = 0
    n=len(points[0])-1
    M = len(points)
    ans = 0
    for i in range(M):
        x = points[i,0:n]
        y = points[i,n]
        ans = 0
        for j in range(n):
                ans += (m[j] * x[j])
        total_cost += (1/M) * ((y - ans ) ** 2)
    return total_cost
```

```python
def pred(m):
    points = np.loadtxt("gdtest1.csv",delimiter = ",")
    n=len(points[0])
    M = len(points)
    ypred = np.zeros(M)
    for i in range(M):
        x = points[i,0:n+1]
        for j in range(n ):
            ypred[i] += (m[j] * x[j]) 
    return ypred
```

```python
def run():
    points = np.loadtxt("gdtrain1.csv",delimiter = ",")
    learning_rate = 0.0025
    no_iteration = 75
    m = gd(points, learning_rate, no_iteration)
    y = pred(m)
    print(y)
    
```
